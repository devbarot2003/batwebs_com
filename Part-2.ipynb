```python
#displaying the values of Id and Species where sepal length greater than 6 in cm.

df2.loc [df2['SepalLengthCm']>6.0, ['Id', 'Species']]
```

```python
#four values of petal length having 5.0 cm

df2.loc[df2['PetalLengthCm'] == 5.0]
```

```python
#5.0 getting replaced by 5.1

df2.loc[df2['PetalLengthCm']==5.0, ['PetalLengthCm']] = 5.1
```

```python
#verifying that the data got replaced by 5.1 or not

df2.loc[df2['PetalLengthCm'] == 5.0]
```

```python
#successfully the data got replaced by 5.1

df2.loc[df2['PetalLengthCm'] == 5.1]
```
